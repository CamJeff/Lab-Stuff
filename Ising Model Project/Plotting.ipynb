```python
%matplotlib inline
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from google.colab import files

#----------------------------------------------------------------------
#  DOCUMENTS AND STUFF
#----------------------------------------------------------------------
def load_csv(filename):
    data = np.genfromtxt(filename, delimiter=',', names=True)
    return data['Temperature'], data['Value'], data['Error']

T, E, E_err = load_csv("energy_vs_temperature.csv")
_, M, M_err = load_csv("magnetization_vs_temperature.csv")
_, C, C_err = load_csv("specific_heat_vs_temperature.csv")
_, X, X_err = load_csv("susceptibility_vs_temperature.csv")


_, E_res, _ = load_csv("energy_residuals.csv")
_, M_res, _ = load_csv("magnetization_residuals.csv")
_, C_res, _ = load_csv("specific_heat_residuals.csv")
_, X_res, _ = load_csv("susceptibility_residuals.csv")

#----------------------------------------------------------------------
#  PLOTTING
#----------------------------------------------------------------------
plt.style.use('seaborn-v0_8-talk')
f = plt.figure(figsize=(20, 16))

# Energy Plot
ax1 = f.add_subplot(3, 2, 1)
ax1.errorbar(T, E, yerr=E_err, fmt='o', color='Red', markersize=6,
            capsize=3, capthick=1, elinewidth=1, label='Energy')
ax1.set_xlabel("Temperature (T)", fontsize=25)
ax1.set_ylabel("Energy", fontsize=25)
ax1.grid(True, alpha=0.3)

# Energy Residuals
ax2 = f.add_subplot(3, 2, 2)
ax2.scatter(T, E_res, color='Red', s=30, alpha=0.7)
ax2.axhline(0, color='gray', linestyle='--')
ax2.set_xlabel("Temperature (T)", fontsize=25)
ax2.set_ylabel("Energy Residuals", fontsize=25)
ax2.grid(True, alpha=0.3)

# Magnetization Plot
ax3 = f.add_subplot(3, 2, 3)
ax3.errorbar(T, abs(M), yerr=M_err, fmt='o', color='Blue', markersize=6,
            capsize=3, capthick=1, elinewidth=1, label='Magnetization')
ax3.set_xlabel("Temperature (T)", fontsize=25)
ax3.set_ylabel("Magnetization", fontsize=25)
ax3.grid(True, alpha=0.3)

# Magnetization Residuals
ax4 = f.add_subplot(3, 2, 4)
ax4.scatter(T, M_res, color='Blue', s=30, alpha=0.7)
ax4.axhline(0, color='gray', linestyle='--')
ax4.set_xlabel("Temperature (T)", fontsize=25)
ax4.set_ylabel("Magnetization Residuals", fontsize=25)
ax4.grid(True, alpha=0.3)

# Specific Heat Plot
ax5 = f.add_subplot(3, 2, 5)
ax5.errorbar(T, C, yerr=C_err, fmt='o', color='Green', markersize=6,
            capsize=3, capthick=1, elinewidth=1, label='Specific Heat')
ax5.set_xlabel("Temperature (T)", fontsize=25)
ax5.set_ylabel("Specific Heat", fontsize=25)
ax5.grid(True, alpha=0.3)

# Specific Heat Residuals
ax6 = f.add_subplot(3, 2, 6)
ax6.scatter(T, C_res, color='Green', s=30, alpha=0.7)
ax6.axhline(0, color='gray', linestyle='--')
ax6.set_xlabel("Temperature (T)", fontsize=25)
ax6.set_ylabel("Specific Heat Residuals", fontsize=25)
ax6.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Additional plot for Susceptibility
f2 = plt.figure(figsize=(14, 6))
ax7 = f2.add_subplot(1, 2, 1)
ax7.errorbar(T, X, yerr=X_err, fmt='o', color='Purple', markersize=6,
            capsize=3, capthick=1, elinewidth=1, label='Susceptibility')
ax7.set_xlabel("Temperature (T)", fontsize=25)
ax7.set_ylabel("Susceptibility", fontsize=25)
ax7.grid(True, alpha=0.3)

ax8 = f2.add_subplot(1, 2, 2)
ax8.scatter(T, X_res, color='Purple', s=30, alpha=0.7)
ax8.axhline(0, color='gray', linestyle='--')
ax8.set_xlabel("Temperature (T)", fontsize=25)
ax8.set_ylabel("Susceptibility Residuals", fontsize=25)
ax8.grid(True, alpha=0.3)

plt.tight_layout()

# This safes an svg for the magnetic sus graph. Move above to safe the three above graphs
plt.savefig('MatrixState.svg', format='svg')
files.download('MatrixState.svg')
plt.show()
```
